From the above data:

1. Read the data with pandas and find features and target variables
2. Find the target variable
3. Plot a graph between features and target
4. Normalize the give data[0 to 1]

```python
import pandas as pd
```

```python
sal_d = pd.read_csv('Salary_Data.csv')
sal_d = sal_d.fillna(0)
sal_d
```

```python
sal = sal_d['Salary']#Target
exp = sal_d.drop('Salary',axis=1)#Features
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
plt.scatter(exp,sal)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Experience VS Salary')
plt.show()
```

```python
Norm_data = (sal_d-sal_d.min())/(sal_d.max()-sal_d.min())
Norm_data
```
