## **Question-1**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('train.csv')
data = data.fillna(0) #Null values
y = data['price_range']
y
```

```python
x = data.drop(['price_range'], axis=1)
x
```

```python
ND = (x-x.min())/(x.max()-x.min())
ND
```

```python
x_train , x_test , y_train , y_test = train_test_split(ND,y,test_size=0.2,random_state=30)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```
